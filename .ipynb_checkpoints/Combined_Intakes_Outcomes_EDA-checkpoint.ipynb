```python
import pandas as pd
```

```python
intakes = pd.read_csv('Cleaned_Intakes.csv',index_col=0)
outcomes = pd.read_csv('Cleaned_Outcomes.csv',index_col=0)
```

```python
merged_outer = intakes.merge(outcomes,how='outer',on='Animal ID',suffixes=('_intakes','_outcomes'))
merged_outer.head()
```

```python
merged_outer.info()
```

```python
merged_inner = intakes.merge(outcomes,how='inner',on='Animal ID')
merged_inner.head()
```

```python
merged_inner.shape[0]
```

```python
intakes
```

```python
print("\n")
print("Total cases of intakes "+ str(intakes.shape[0]) + " and total cases of outcomes " + str(outcomes.shape[0]))
print("\n")
print("The number of casee both intake and outcome happened " + str(merged_inner.shape[0]))
print("\n")
print("The number of intake cases without outcomes" + str(intakes.shape[0] - merged_inner.shape[0] ))
print("\n")
print("The number of outcome cases without intake" + str(outcomes.shape[0] - merged_inner.shape[0]))
print("\n")
```

```python
test_intakes = intakes[intakes.duplicated('Animal ID',keep=False)].sort_values(['Animal ID'])
test_intakes.head(5)
```

```python
test_outcomes = outcomes[outcomes.duplicated('Animal ID',keep=False)].sort_values(['Animal ID','DateTime'])
test_outcomes.head(5)
```

```python
merged1 = test_intakes.merge(test_outcomes,how='inner',on='Animal ID')
merged1.shape[0]
```
